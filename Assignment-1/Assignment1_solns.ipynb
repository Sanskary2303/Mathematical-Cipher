```python
#2a)

def Enc(m, k):
    encrypted_message = ""
    for char in m:
        if char.isalpha():  # Check if the character is a letter
            shift = k % 26  # Ensure shift is within 0-25
            if char.islower():
                encrypted_char = chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
            elif char.isupper():
                encrypted_char = chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
            encrypted_message += encrypted_char
        else:
            encrypted_message += char  # Non-alphabetic characters remain unchanged
    return encrypted_message

# Encrypted messages
message1 = "iitk is better than iitd and iitb"
key1 = 9
encrypted_message1 = Enc(message1, key1)

message2 = "lets learn cryptography"
key2 = 25
encrypted_message2 = Enc(message2, key2)

print("Encrypted message 1:", encrypted_message1)
print("Encrypted message 2:", encrypted_message2)

```

```python
#2b)

import string

def decrypt(c, shift):
    decrypted_text = []
    for char in c:
        if char in string.ascii_lowercase:
            decrypted_text.append(chr((ord(char) - shift - ord('a')) % 26 + ord('a')))
        elif char in string.ascii_uppercase:
            decrypted_text.append(chr((ord(char) - shift - ord('A')) % 26 + ord('A')))
        else:
            decrypted_text.append(char)
    return ''.join(decrypted_text)

def find_most_probable_message(c):
    common_words = {"the", "be", "to", "of", "and", "a", "in", "that", "have", "I","it"}
    possible_messages = []
    for shift in range(26):
        decrypted_message = decrypt(c, shift)
        words = decrypted_message.split()
        common_word_count = sum(1 for word in words if word in common_words)
        possible_messages.append((decrypted_message, common_word_count))
    
    # Find the message with the maximum number of common words
    most_probable_message = max(possible_messages, key=lambda x: x[1])
    return most_probable_message[0]

# Encrypted messages
ciphertext1 = 'bm ptl wtfg xtlr tztbg'
ciphertext2 = 'rc fjb mjvw njbh'

# Print all possible messages and the most probable one
print("All possible messages for (i):")
for shift in range(26):
    print(f"Shift {shift}: {decrypt(ciphertext1, shift)}")
print("Most probable message for (i):", find_most_probable_message(ciphertext1))

print("\nAll possible messages for (ii):")
for shift in range(26):
    print(f"Shift {shift}: {decrypt(ciphertext2, shift)}")
print("Most probable message for (ii):", find_most_probable_message(ciphertext2))
```
